```python
def extract_users(info):
    '''
    input:
        info: data['np2users']
    output:
        list_user, user2kw
    '''
    l_user, user2kw = [], []
    for ii in info:
        lus = info[ii]
        for u in lus:
            if u not in l_user:
                l_user.append(u)
                user2kw.append([])
            idx = l_user.index(u)
            user2kw[idx].append(ii)
    return l_user, user2kw
```
